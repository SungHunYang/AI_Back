```python
import numpy as np
import matplotlib.pyplot as plt

function_x = np.linspace(-3,3,100)
function_y = function_x**2

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(function_x,function_y)

x = 2
Ir = 0.1 # 계수 알파
# learning rate가 너무 커지면 오류가 발생할 수 있고,(발산하게 된다.)
# learning rate는 커질수록 간격이 넓어지고, 작을수록 간격이 좁아진다.

for _ in range(10):
  dy_dx = 2*x
  x = x - (Ir*dy_dx) # x를 업데이트하는 식
  y=x**2

  ax.scatter(x,y,color='red', s= 100) # 점점 0에 다가오는것을 확인 할 수있다.
```
